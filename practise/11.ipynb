```python
#11-1
import numpy as np
import cv2

# 固定亂數種子以便重現
np.random.seed(0)

# 產生隨機灰階影像 (512x512)
o = np.random.randint(0, 256, size=[10,10], dtype=np.uint8)

# 兩組 Canny 閾值
r1 = cv2.Canny(o, 150, 200)
r2 = cv2.Canny(o, 50, 100)

print("original image:\n",o)
print("Canny edges (150, 200):\n",r1)
print("Canny edges (50, 100):\n",r2)
```

```python
# 11-2 Case Study
import os
import cv2
import numpy as np
from matplotlib import pyplot as plt

img_path = r"C:/img/lena.bmp"

if not os.path.exists(img_path):
    print(f"File not found: {img_path}")
else:
    # 以灰階讀取
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        print(f"Failed to read image: {img_path}")
    else:

        # 兩組 Canny 閾值
        edges_150_200 = cv2.Canny(img, 150, 200)
        edges_50_100 = cv2.Canny(img, 50, 100)

        # 視覺化
        fig, axs = plt.subplots(1, 3, figsize=(12, 4))
        axs[0].imshow(img, cmap='gray')
        axs[0].set_title('Original (Lena)')
        axs[0].axis('off')

        axs[1].imshow(edges_150_200, cmap='gray')
        axs[1].set_title('Canny (150, 200) + Gaussian(5,1.0)')
        axs[1].axis('off')

        axs[2].imshow(edges_50_100, cmap='gray')
        axs[2].set_title('Canny (50, 100) + Gaussian(5,1.0)')
        axs[2].axis('off')

        plt.tight_layout()
        plt.show()

        # 統計邊緣像素數
        print(f"Edges (150,200): {np.count_nonzero(edges_150_200)} pixels")
        print(f"Edges (50,100): {np.count_nonzero(edges_50_100)} pixels")
```
