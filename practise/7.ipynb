```python
# 練習 7-3 Shift
import cv2
import numpy as np

# 讀取圖片
img = cv2.imread("C:/img/lena512.bmp")
cv2.imshow("Original", img)

# 圖片大小
height, width = img.shape[:2]

# 平移量
x1 = 100   # 向右
y1 = 200   # 向下

# 平移矩陣
M = np.float32([[1, 0, x1], [0, 1, y1]])

# 平移圖片
move = cv2.warpAffine(img, M, (width, height))

# 顯示平移後圖片
cv2.imshow("Shifted", move)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 練習 7-4 Rotation
import cv2
import numpy as np

# 讀取圖片
img = cv2.imread("C:/img/lena512.bmp")
cv2.imshow("Original", img)
height, width = img.shape[:2]
M=cv2.getRotationMatrix2D((width/2,height/2),45,0.6)
rotate=cv2.warpAffine(img,M,(width,height))

cv2.imshow("Shifted", rotate)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 練習 7-5
import cv2
import numpy as np

# 隨機產生 4x5 的矩陣
img = np.random.randint(0, 256, size=[4,5], dtype=np.uint8)

rows, cols = img.shape

# 設定對應矩陣
mapx = np.ones(img.shape, np.float32) * 2 
mapy = np.ones(img.shape, np.float32) * 3  

# 使用 remap 重新映射
rst = cv2.remap(img, mapx, mapy, cv2.INTER_LINEAR)
print("img\n",img,'\n')
print("mapx\n",mapx,'\n')
print("mapy\n",mapy,'\n')
print("rst\n",rst)
```

```python
# 練習 7-6 Re-Mapping and Copy
import cv2
import numpy as np

# 隨機產生 4x5 的矩陣
img = np.random.randint(0, 256, size=[4,5], dtype=np.uint8)

rows, cols = img.shape

# 初始化 mapx 和 mapy
mapx = np.zeros((rows, cols), dtype=np.float32)
mapy = np.zeros((rows, cols), dtype=np.float32)

# 用 for 迴圈生成對應矩陣
for i in range(rows):
    for j in range(cols):
        mapx[i, j] = j  # 對應 x 座標
        mapy[i, j] = i  # 對應 y 座標

# 使用 remap
rst = cv2.remap(img, mapx, mapy, cv2.INTER_LINEAR)

print("img\n", img, '\n')
print("mapx\n", mapx, '\n')
print("mapy\n", mapy, '\n')
print("rst\n", rst)
```

```python
# 練習 7-6 Re-Mapping and Copy
import cv2
import numpy as np

# 隨機產生 4x5 的矩陣
img = np.random.randint(0, 256, size=[4,5], dtype=np.uint8)

rows, cols = img.shape

# 初始化 mapx 和 mapy
mapx = np.zeros((rows, cols), dtype=np.float32)
mapy = np.zeros((rows, cols), dtype=np.float32)

# 用 for 迴圈生成對應矩陣
for i in range(rows):
    for j in range(cols):
        mapx[i, j] = cols-j-1  
        mapy[i, j] = rows-i-1 

# 使用 remap
rst = cv2.remap(img, mapx, mapy, cv2.INTER_LINEAR)

print("img\n", img, '\n')
print("mapx\n", mapx, '\n')
print("mapy\n", mapy, '\n')
print("rst\n", rst)
```
