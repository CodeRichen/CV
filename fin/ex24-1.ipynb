
| 案例   | Precision         | Recall            | F1-score         | Dice             | IoU              |
|--------|-------------------|-------------------|------------------|------------------|------------------|
| Case 1 | $1$ (10/10)       | $0.1$ (10/100)    | $0.182$          | $1/5.5\approx0.18$ | $1/11\approx0.09$ |
| Case 2 | $0.5$ (5/10)      | $0.05$ (5/100)    | $0.091$          | $1/11\approx0.09$  | $5/110\approx0.045$ |
| Case 3 | $0.05$ (5/100)    | $0.5$ (5/10)      | $0.091$          | $1/11\approx0.09$  | $5/110\approx0.045$ |
| Case 4 | $0.1$ (10/100)    | $1$ (10/10)       | $0.182$          | $1/5.5\approx0.18$ | $1/11\approx0.09$ |

- **Precision (精確率)**：TP / (TP + FP)
- **Recall (召回率)**：TP / (TP + FN)
- **F1-score**：$2 / (1/Precision + 1/Recall)$
- **Dice**：$2 \times TP / (2 \times TP + FP + FN)$
- **IoU (交集比聯集)**：$TP / (TP + FP + FN)$
