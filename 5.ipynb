```python
# 練習5-1
import numpy as np
import cv2
# 隨機產生 4x5 影像
img = np.random.randint(0, 256, (4, 5), dtype=np.uint8)
print("原圖：\n", img)

# 建立空矩陣
binary_loop = np.zeros_like(img)

# 門檻值
th = 127

# 利用雙層迴圈做二值化
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if img[i, j] > th:
            binary_loop[i, j] = 255
        else:
            binary_loop[i, j] = 0

print("\n迴圈版二值化結果：\n", binary_loop)

t,rst=cv2.threshold(img,127,255,cv2.THRESH_BINARY)

print("\nmask 版二值化結果：\n", rst)
```

```python
# 練習5-2 Binary Theresholing
import cv2
import numpy as np

# 讀取灰階圖片
img = cv2.imread(r"C:\img\lena512.bmp", 0)
cv2.imshow("Original", img)

# 設定門檻值
th = 127

# --- 方法 1：用迴圈 ---
binary_loop = np.zeros_like(img)
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if img[i, j] > th:
            binary_loop[i, j] = 255
        else:
            binary_loop[i, j] = 0

cv2.imshow("Binary by Loop", binary_loop)

# --- 方法 2：用 mask ---

t,rst=cv2.threshold(img,127,255,cv2.THRESH_BINARY)
cv2.imshow("Binary by Mask", rst)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 練習5-4 Threshold Truncate
import cv2
import numpy as np

# 讀入灰階圖像
img = np.random.randint(0, 256, (4, 5), dtype=np.uint8)

TH = 127  # 閾值

# 建立輸出矩陣（與原圖同大小）
rst = np.zeros_like(img)

# 手工實作 TRUNC 閾值化
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if img[i, j] > TH:
            rst[i, j] = TH
        else:
            rst[i, j] = img[i, j]
_, rst2 = cv2.threshold(img, TH, 255, cv2.THRESH_TRUNC)
print("原本:\n",img)
print()
print("手工打造:\n",rst)
print()
print("函式:\n",rst2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 練習5-6 Threshold to 0 (Inverse)
import cv2
import numpy as np

# 讀取灰階影像
img = np.random.randint(0, 256, (4, 5), dtype=np.uint8)

# 門檻值
TH = 127

# ---- OpenCV 版本 ----
_, rst1 = cv2.threshold(img, TH, 255, cv2.THRESH_TOZERO_INV)

# ---- 手工打造版本 ----
rst2 = img.copy()
rst2[img > TH] = 0   # 大於門檻改為0，其餘保留

# ---- 顯示結果 ----
print("原本:\n",img)
print()
print("手工打造:\n",rst1)
print()
print("函式:\n",rst2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 練習5-8 Adaptive Threshold
import cv2

# 讀灰階圖片
img = cv2.imread(r"C:\img\caption.jpg", cv2.IMREAD_GRAYSCALE)
t1,thd=cv2.threshold(img,127,255,cv2.THRESH_BINARY)
# --- 自適應閾值：平均值法 ---
binary_mean = cv2.adaptiveThreshold(
    img,                  # 原圖
    255,                  # 最大值
    cv2.ADAPTIVE_THRESH_MEAN_C,  # 自適應方法
    cv2.THRESH_BINARY,    # 閾值類型
    3,                    # blockSize (3x3 區塊)
    5                     # C: 偏移量
)

# --- 自適應閾值：高斯法 ---
binary_gaussian = cv2.adaptiveThreshold(
    img,
    255,
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY,
    3,
    5
)

# --- 顯示 ---
cv2.imshow("Original", img)
cv2.imshow("thd", thd)
cv2.imshow("Adaptive Mean", binary_mean)
cv2.imshow("Adaptive Gaussian", binary_gaussian)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 練習5-11
import cv2
import numpy as np

# 1️⃣ 讀彩色圖片
img = cv2.imread(r"C:\img\flower.jpg")

# 2️⃣ 轉 HSV
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# 3️⃣ 粗略黃色遮罩 (H: 20~35, S & V 可調整)
lower_yellow = np.array([18, 100, 100], dtype=np.uint8)
upper_yellow = np.array([28, 255, 255], dtype=np.uint8)
mask_yellow = cv2.inRange(hsv, lower_yellow, upper_yellow)

# 4️⃣ 取 V 通道 (亮度)
v_channel = hsv[:, :, 2]

# 5️⃣ 只保留黃色區域亮度，其餘設 0
v_masked = cv2.bitwise_and(v_channel, v_channel, mask=mask_yellow)

# 6️⃣ Otsu 自動二值化，分割病蟲害
t, otsu_mask = cv2.threshold(v_masked, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# 7️⃣ 用 Otsu 遮罩套用原彩色圖
result = cv2.bitwise_and(img, img, mask=otsu_mask)

# 8️⃣ 顯示結果
cv2.imshow("Original", img)
# cv2.imshow("Yellow Mask", mask_yellow)       # 粗略黃色遮罩
# cv2.imshow("V Channel Masked", v_masked)     # V 通道亮度
# cv2.imshow("Otsu Mask", otsu_mask)          # Otsu 二值化結果
cv2.imshow("Segmented Yellow", result)      # 彩色黃色病蟲害
cv2.waitKey(0)
cv2.destroyAllWindows()
```
