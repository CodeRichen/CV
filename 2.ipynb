```python
#練習2-10 Copy of ROI
import cv2

# 讀取影像
img = cv2.imread(r"C:\img\lena512.bmp", cv2.IMREAD_UNCHANGED)
bg = cv2.imread(r"C:\img\dollar.bmp", 0)

# 取出 face
face = img[210:400, 250:350]

# 縮小 90%
new_width = int(face.shape[1] * 0.95)
new_height = int(face.shape[0] * 0.95)
face = cv2.resize(face, (new_width, new_height))

# 計算貼圖索引 (一定要整數)
y1 = 210 - 50
y2 = y1 + new_height
x1 = 253 - 50
x2 = x1 + new_width

bg[y1:y2, x1:x2] = face  # 貼上

cv2.imshow("result", bg)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 用 OpenCV 顯示圖片並用滑鼠點擊座標
import cv2
coords = []

def mouse_callback(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:  # 滑鼠左鍵點擊
        print(f"Clicked: ({x}, {y})")
        coords.append((x, y))

img = cv2.imread(r"C:\img\lena.bmp")
cv2.imshow("img", img)
cv2.setMouseCallback("img", mouse_callback)
cv2.waitKey(0)
cv2.destroyAllWindows()

print("Collected coordinates:", coords)
# [(571, 550), (634, 550), (571, 577), (634, 579)]
# [(194, 208), (258, 205), (194, 235), (259, 235)]

#  [(123, 117), (176, 120), (117, 189), (173, 191)]
# [(123, 120), (180, 125), (120, 141), (178, 143)]
```

```python
#Exercise 練習2-11
import cv2

# 讀取兩張車圖片
car1 = cv2.imread(r"C:\img\car01.jpg")
car2 = cv2.imread(r"C:\img\car02.jpg")
cv2.imshow("Car1 before swapped", car1)
cv2.imshow("Car2 before swapped", car2)
# 假設車牌座標 (左上角x, 左上角y, 右下角x, 右下角y)
plate1_coords = (571, 550, 634, 579)  # car1 車牌區域
plate2_coords = (194, 208, 259, 235)  # car2 車牌區域

# 取 ROI
x1, y1, x2, y2 = plate1_coords
plate1 = car1[y1:y2, x1:x2].copy()  # 用 copy 避免引用問題

x3, y3, x4, y4 = plate2_coords
plate2 = car2[y3:y4, x3:x4].copy()

# 對調
car1[y1:y2, x1:x2] = cv2.resize(plate2, (x2-x1, y2-y1))
car2[y3:y4, x3:x4] = cv2.resize(plate1, (x4-x3, y4-y3))

# 顯示結果
cv2.imshow("Car1 swapped", car1)
cv2.imshow("Car2 swapped", car2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 練習2-12 Channel Split by Index 令解
import cv2
import numpy as np
lena=cv2.imread('C:/img/lenacolor.png')
cv2.imshow('lena1',lena)
b=lena[:,:,0]
g=lena[:,:,1]
r=lena[:,:,2]
cv2.imshow('b',b)
cv2.imshow('g',g)
cv2.imshow('r',r)
img=lena
# 拆開 BGR
B, G, R = cv2.split(img)

# 調整通道
R = cv2.add(R, 55)      # 紅色加強
G = cv2.add(G, 35)      # 綠色稍微加強
B = cv2.subtract(B, 20) # 藍色減弱

# 合併回去
lenab0 = cv2.merge([B, G, R])

cv2.imshow('lenab0',lenab0)

img=lena
rad = np.full(img.shape, (0, 0, 255), dtype=np.uint8)
lenab0g0 = cv2.addWeighted(img, 0.4, rad, 0.6, 0)

cv2.imshow('lenab0g0',lenab0g0)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 練習2-14 Channel Combination
import cv2
lena=cv2.imread('C:/img/lenacolor.png')
b,g,r=cv2.split(lena)
bgr=cv2.merge([b,g,r])
rgb=cv2.merge([r,g,b])
cv2.imshow("lena",lena)
cv2.imshow("bgr",bgr)
cv2.imshow("rgb",rgb)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Exercise 練習2-15 試用python程式比較bgr，rgb與原圖是否相同?

import cv2
import numpy as np

lena = cv2.imread('C:/img/lenacolor.png')
b,g,r = cv2.split(lena)
bgr = cv2.merge([b,g,r])  # 跟原圖相同
rgb = cv2.merge([r,g,b])  # 跟原圖不同

# 判斷是否相同
print("lena == bgr ?", np.array_equal(lena, bgr))
print("lena == rgb ?", np.array_equal(lena, rgb))
```
